## Problem Statement

Students from different cities from the state of Maharashtra had applied for the Cloud
Counselage Internship Program. We have the dataset of consisting information of all the
students. Using this data we want to get more insights and draw out more meaningful
conclusions. Interns are expected to build a data visualization model and find the best data
segmentation model using the student’s dataset. Following are the tasks interns need to
perform :

#### 1. Interns need to preprocess the data for missing values, unknown values, encoding categorical values.

#### 2. Create a data visualization model to build graphs from the dataset answering the following questions:
        a. The number of students applied to different technologies. 
        b. The number of students applied for Data Science who knew ‘’Python” and who didn’t. 
        c. The different ways students learned about this program. 
        d. Students who are in the fourth year and have a CGPA greater than 8.0. 
        e. Students who applied for Digital Marketing with verbal and written communication score greater than 8.
        f. Year-wise and area of study wise classification of students. 
        g. City and college wise classification of students. 
        h. Plot the relationship between the CGPA and the target variable. 
        i. Plot the relationship between the Area of Interest and the target variable. 
        j. Plot the relationship between the year of study, major, and the target variable.
        
        
#### 3. Identify the best binary classifier to classify data into “eligible/1” and “not eligible/0”.

### Dataset :-

https://cloudcounselage24.bitrix24.com/disk/showFile/36677/?&ncc=1&ts=1590841589&filename=DS_DATESET.csv

### Input

Path of the input dataset file

### Output

For the data classification, In output print accuracy of the best model identified.

For the data visualization model, in output code need to create a pdf file of all the graphs
generated by the model.

## Data Visualization

#### Create a data visualization model to build graphs from the dataset answering the following questions:

    a. The number of students applied to different technologies.
    b. The number of students applied for Data Science who knew ‘’Python” and who didn’t. 
    c. The different ways students learned about this program. 
    d. Students who are in the fourth year and have a CGPA greater than 8.0. 
    e. Students who applied for Digital Marketing with verbal and written communication score greater than 8. 
    f. Year-wise and area of study wise classification of students. 
    g. City and college wise classification of students. 
    h. Plot the relationship between the CGPA and the target variable. 
    i. Plot the relationship between the Area of Interest and the target variable. 
    j. Plot the relationship between the year of study, major, and the target variable.


### a. The number of students applied to different technologies.

```python
figure(figsize=(10, 10))

countplot(data=df,
          y=df['Areas of interest'],
          order = df['Areas of interest'].value_counts().index,
          palette =sns.color_palette("Blues_d", n_colors=df['Areas of interest'].nunique()))

plt.xlabel("Number of Students", labelpad=15, fontsize = 14)
plt.ylabel("Areas of interest", labelpad=15, fontsize = 14)
plt.title("The number of students applied to different technologies",  y=1.03, fontsize = 16);
```

### b. The number of students applied for Data Science who knew ‘’Python” and who didn’t.

```python
df_ds = df.loc[df['Areas of interest']=='Data Science ']
df_ds_py = df_ds.loc[df['Programming Language Known other than Java (one major)']=='Python']

figure(figsize=(8,8))

values=[df_ds.shape[0], df_ds_py.shape[0]]
explode=[0,0,0.1,0,0]

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}\n({:.2f}%)".format(absolute, pct)


plt.pie(values,
        labels = ["Doesn't know Python", 'Knows Python'],
        autopct=lambda pct: func(pct, values),
        explode=[0,0.1],
        textprops=dict(size=14))
plt.legend(bbox_to_anchor=(1, 0.9, 0, 0), fontsize=13)
plt.title('The number of students applied for "Data Science" who knew "Python" and who didn’t', y=1.0, fontsize = 16);
```

### c. The different ways students learned about this program.

```python
figure(figsize=(10, 10))

countplot(data=df,
          y=df['How Did You Hear About This Internship?'],
          order = df['How Did You Hear About This Internship?'].value_counts().index,
          palette =sns.color_palette("Blues_d", n_colors=df['How Did You Hear About This Internship?'].nunique()))

plt.xlabel("Number of Students", labelpad=15, fontsize = 14)
plt.ylabel("How Did You Hear About This Internship?", labelpad=15, fontsize = 14)
plt.title("The number of students applied to different technologies",  y=1.03, fontsize = 16);
```

### d. Students who are in the fourth year and have a CGPA greater than 8.0.

```python
df_d = df.loc[(df['Which-year are you studying in?'] == 'Fourth-year') & 
              (df['CGPA/ percentage'] > 8.0)]

figure(figsize = (15,7))
sns.distplot(a=df_d['CGPA/ percentage'],hist= True,
             kde_kws={"color": "red","lw" : "4"},
             hist_kws={"color": "orange", "label": "CGPA density"})
```

### e. Students who applied for Digital Marketing with verbal and written communication score greater than 8.

```python
df_e = df.loc[(df['Areas of interest'] == 'Digital Marketing ') &
              (df['Rate your verbal communication skills [1-10]'] > 8) &
              (df['Rate your written communication skills [1-10]'] > 8)]
```

```python
fig,ax = plt.subplots(figsize = (15,7))
ax.set_ylabel('Number of Students')
ax.set_xlabel('pointer rating')

comm_counts = df_e['Rate your written communication skills [1-10]'].value_counts()
verbal_counts = df_e['Rate your verbal communication skills [1-10]'].value_counts()
sns.lineplot(x= comm_counts.index, y = comm_counts.values,ax = ax,label = 'Written Communication rating')
sns.lineplot(x = verbal_counts.index, y = verbal_counts.values,ax =ax,label  = 'Verbal communication rating')
```

### f. Year-wise and area of study wise classification of students.

```python
df_f = df[['Which-year are you studying in?', 'Major/Area of Study']]
fig,ax = plt.subplots(figsize = (20,10))
df_f.groupby(['Which-year are you studying in?', 'Major/Area of Study',]).size().unstack().plot.barh(ax =ax,rot = 0);
```

```python
df.columns
```

### g. City and college wise classification of students.

```python
fig,ax = plt.subplots(figsize = (25,15))
pd.crosstab(df['City'],df['College name']).plot.bar(ax =ax,rot = 0)
```

### h. Plot the relationship between the CGPA and the target variable.

```python
figure(figsize = (15,10))
sns.violinplot(y="CGPA/ percentage", x="Label",data=df);
```

### i. Plot the relationship between the Area of Interest and the target variable.

```python
figure(figsize = (15,10))
sns.violinplot(y=df['Major/Area of Study'].index,
               x=df['Label'],
               data=df);
```

### j. Plot the relationship between the year of study, major, and the target variable.

```python
major_df= df.groupby(['Major/Area of Study','Which-year are you studying in?','Label'])['Major/Area of Study'].size()[lambda x: x < 1000]

fig,ax = plt.subplots(figsize = (15,7) )
major_df = major_df.to_frame()
major_df.unstack().plot.bar(ax =ax)
```

## 3. Identify the best binary classifier to classify data into “eligible/1” and “not eligible/0”.

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df = df.apply(LabelEncoder().fit_transform)
df
```

```python
from sklearn.model_selection import train_test_split

X = df[['First Name', 'Last Name', 'City', 'State', 'Zip Code',
       'DOB [DD/MM/YYYY]', 'Age', 'Gender', 'Email Address', 'Contact Number',
       'Emergency Contact Number', 'College name', 'University Name', 'Degree',
       'Major/Area of Study', 'Course Type', 'Which-year are you studying in?',
       'CGPA/ percentage', 'Expected Graduation-year', 'Areas of interest',
       'Current Employment Status', 'Have you worked core Java',
       'Programming Language Known other than Java (one major)',
       'Have you worked on MySQL or Oracle database',
       'Have you studied OOP Concepts',
       'Rate your written communication skills [1-10]',
       'Rate your verbal communication skills [1-10]',
       'How Did You Hear About This Internship?']]
```

```python
y = df.Label
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 1)
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 500)
rf.fit(X_train,y_train)
```

```python
RandomForestClassifier(n_estimators=500)
```

```python
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix,plot_confusion_matrix,r2_score
import itertools
y_pred = rf.predict(X_test)

score = accuracy_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)


print('accuracy  =',score, 'f1 = ', f1)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
plot_confusion_matrix(rf,X,y)
```

```python
temp = list()
for index, row in df_new.iterrows():
    calnum = (
                 (row['CGPA/ percentage'] / 10) \
                 * ((row["Rate your verbal communication skills [1-10]"] / 10) \
                 + (row['Rate your written communication skills [1-10]'] / 10)) \
                 # + 10 ) \
                 ** (row['Which-year are you studying in?'] / 1000)
        )
    
    temp.append(calnum)
    
df2 = df_new.assign(temp = temp)
corr = df2[['temp', 'Label']].corr()
corr
```
